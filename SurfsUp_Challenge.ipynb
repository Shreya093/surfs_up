```python
# Dependencies
import numpy as np
import pandas as pd

# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)

# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

## D1: Determine the Summary Statistics for June

```python
# 1. Import the sqlalchemy extract function.
from sqlalchemy import extract

# 2. Write a query that filters the Measurement table to retrieve the temperatures for the month of June. 
june_temp = session.query(Measurement.date, Measurement.tobs).\
            filter(extract('month', Measurement.date) == 6).all()

june_temp
```

```python
#  3. Convert the June temperatures to a list.
results = session.query(Measurement.tobs).\
            filter(extract('month', Measurement.date) == 6).all()

june_temp_list = list(np.ravel(results))

june_temp_list
```

```python
# 4. Create a DataFrame from the list of temperatures for the month of June. 
june_temp_df = pd.DataFrame(june_temp_list, columns = ['June Tobs'])

june_temp_df
```

```python
# 5. Calculate and print out the summary statistics for the June temperature DataFrame.
june_temp_df.describe()
```

## D2: Determine the Summary Statistics for December

```python
# 6. Write a query that filters the Measurement table to retrieve the temperatures for the month of December.
december_temp = session.query(Measurement.date, Measurement.tobs).\
                filter(extract('month', Measurement.date) == 12).all()

december_temp
```

```python
# 7. Convert the December temperatures to a list.
results = session.query(Measurement.tobs).\
          filter(extract('month', Measurement.date) == 12).all()

december_temp_list = list(np.ravel(results))

december_temp_list
```

```python
# 8. Create a DataFrame from the list of temperatures for the month of December. 
december_temp_df = pd.DataFrame(december_temp_list, columns = ['December Tobs'])

december_temp_df
```

```python
# 9. Calculate and print out the summary statistics for the Decemeber temperature DataFrame.
december_temp_df.describe()
```

```python
# Additional Query: 1
# Write a query that filters the Measurement table to retrieve the precipitation for the month of June.
# Convert the list into a Dataframe.

june_prcp = session.query(Measurement.tobs, Measurement.prcp).\
            filter(extract('month', Measurement.date) == 6).all()

june_prcp_df = pd.DataFrame(june_prcp, columns = ['Tobs', 'Prcp'])
june_prcp_df
```

```python
# Generate a summary statistics for the dataframe.
june_prcp_df.describe()
```

```python
# June Precipitation Histogram
june_prcp = session.query(Measurement.prcp).filter(extract('month', Measurement.date)==6).\
            filter(Measurement.station =='USC00519281')
june_prcp_list = [precip.prcp for precip in june_prcp]
juneprcp_df = pd.DataFrame(june_prcp_list, columns=["June Precipitation (in inches)"])
juneprcp_df.plot.hist(bins=5)
```

```python
# Additional Query: 2
# Write a query that filters the Measurement table to retrieve the precipitation for the month of December.
# Convert the list into a Dataframe.

december_prcp = session.query(Measurement.tobs, Measurement.prcp).\
            filter(extract('month', Measurement.date) == 12).all()

december_prcp_df = pd.DataFrame(december_prcp, columns = ['Tobs', 'Prcp'])
december_prcp_df
```

```python
# Generate a summary statistics for the dataframe.
december_prcp_df.describe()
```
